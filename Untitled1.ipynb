```python
#load data 

import pandas as pd

train = pd.read_csv("C:/Users/Aravind/Downloads/w beyound/store_forecasting_data (1) (1)/store_forecasting_data/train.csv")
test = pd.read_csv("C:/Users/Aravind/Downloads/w beyound/store_forecasting_data (1) (1)/store_forecasting_data/test.csv")
stores = pd.read_csv("C:/Users/Aravind/Downloads/w beyound/store_forecasting_data (1) (1)/store_forecasting_data/stores.csv")
oil = pd.read_csv("C:/Users/Aravind/Downloads/w beyound/store_forecasting_data (1) (1)/store_forecasting_data/oil.csv")
holidays = pd.read_csv('C:/Users/Aravind/Downloads/w beyound/store_forecasting_data (1) (1)/store_forecasting_data/holidays_events.csv')
transaction = pd.read_csv('C:/Users/Aravind/Downloads/w beyound/store_forecasting_data (1) (1)/store_forecasting_data/transactions.csv')
```

```python
#handling missing value
#interpolate used when The data is continuous
oil['dcoilwtico'] = oil['dcoilwtico'].interpolate()
```

```python
#transforming time 
train['date'] = pd.to_datetime(train['date'])
test['date'] = pd.to_datetime(test['date'])
oil['date'] = pd.to_datetime(oil['date'])
holidays['date'] = pd.to_datetime(holidays['date'])
transaction['date'] = pd.to_datetime(transaction['date'])
```

```python
train = train.merge(stores, on='store_nbr', how='left')
# Merge with oil.csv
oil['dcoilwtico'] = oil['dcoilwtico'].interpolate()  # Fill missing oil prices
train = train.merge(oil, on='date', how='left')

#  Merge with holidays_events.csv
holidays['is_holiday'] = 1
holiday_flags = holidays[['date', 'is_holiday']].drop_duplicates()
train = train.merge(holiday_flags, on='date', how='left')
train['is_holiday'] = train['is_holiday'].fillna(0)
main_data = train.copy()
```

```python
'''Time-based Features:
  - Extract day, week, month, year, and day of the week.'''

main_data['day'] = main_data['date'].dt.day
main_data['week'] = main_data['date'].dt.isocalendar().week.astype(int)
main_data['month'] = main_data['date'].dt.month
main_data['year'] = main_data['date'].dt.year
main_data['day_of_week'] = main_data['date'].dt.dayofweek  # 0 = Monday

# Identify seasonal month (December = holiday sales peak)
main_data['is_december'] = (main_data['month'] == 12).astype(int)
```

```python
main_data['month'] = main_data['date'].dt.month
main_data['month_name'] = main_data['date'].dt.strftime('%B')  # For readable month names
```

```python
monthly_sales = main_data.groupby('month_name')['sales'].sum().reset_index()
# Optional: Sort by calendar month order
import calendar
monthly_sales['month_num'] = monthly_sales['month_name'].apply(lambda x: list(calendar.month_name).index(x))
monthly_sales = monthly_sales.sort_values('month_num')
```

```python
main_data = main_data.sort_values(['store_nbr', 'family', 'date'])

# Rolling Mean and Std over past 7 and 30 days
main_data['rolling_mean_7'] = main_data.groupby(['store_nbr', 'family'])['sales'].transform(lambda x: x.shift(1).rolling(7).mean())
main_data['rolling_std_7'] = main_data.groupby(['store_nbr', 'family'])['sales'].transform(lambda x: x.shift(1).rolling(7).std())

main_data['rolling_mean_30'] = main_data.groupby(['store_nbr', 'family'])['sales'].transform(lambda x: x.shift(1).rolling(30).mean())
main_data['rolling_std_30'] = main_data.groupby(['store_nbr', 'family'])['sales'].transform(lambda x: x.shift(1).rolling(30).std())
```

```python
# Sort by store and date to apply rolling features properly
main_data = main_data.sort_values(['store_nbr', 'date'])

# 7-day and 30-day rolling mean & std for each store
main_data['rolling_mean_7'] = main_data.groupby('store_nbr')['sales'].transform(lambda x: x.rolling(7).mean())
main_data['rolling_std_7'] = main_data.groupby('store_nbr')['sales'].transform(lambda x: x.rolling(7).std())

main_data['rolling_mean_30'] = main_data.groupby('store_nbr')['sales'].transform(lambda x: x.rolling(30).mean())
main_data['rolling_std_30'] = main_data.groupby('store_nbr')['sales'].transform(lambda x: x.rolling(30).std())
```

```python
# Lagged sales: 1-day, 7-day, 30-day
main_data['lag_1'] = main_data.groupby('store_nbr')['sales'].shift(1)
main_data['lag_7'] = main_data.groupby('store_nbr')['sales'].shift(7)
main_data['lag_30'] = main_data.groupby('store_nbr')['sales'].shift(30)
```

```python
# Avg sales per store type
avg_sales_by_type = df.groupby('type')['sales'].mean().reset_index().rename(columns={'sales': 'avg_sales_by_type'})
df = df.merge(avg_sales_by_type, on='type', how='left')

# Top-selling product families per cluster
top_families = df.groupby(['cluster', 'family'])['sales'].sum().reset_index()
top_families = top_families.sort_values(['cluster', 'sales'], ascending=[True, False])

# Get top 1 per cluster
top_family_per_cluster = top_families.groupby('cluster').first().reset_index().rename(columns={'family': 'top_family'})

# Merge into df
df = df.merge(top_family_per_cluster[['cluster', 'top_family']], on='cluster', how='left')

# Create binary flag
df['is_top_family_in_cluster'] = (df['family'] == df['top_family']).astype(int)
```

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import numpy as np

scaler = MinMaxScaler()
scaled = scaler.fit_transform(daily_sales.values.reshape(-1, 1))

# Create sequences
def create_sequences(data, steps):
    X, y = [], []
    for i in range(steps, len(data)):
        X.append(data[i-steps:i])
        y.append(data[i])
    return np.array(X), np.array(y)

n_steps = 7
X_seq, y_seq = create_sequences(scaled, n_steps)

split_idx = daily_sales.index.get_loc(split_date) - n_steps
X_train, X_test = X_seq[:split_idx], X_seq[split_idx:]
y_train, y_test = y_seq[:split_idx], y_seq[split_idx:]

# LSTM Model
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(n_steps, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=20, verbose=0)

y_pred_lstm = model.predict(X_test)
y_pred_lstm_inv = scaler.inverse_transform(y_pred_lstm)
y_test_inv = scaler.inverse_transform(y_test)

evaluate(y_test_inv, y_pred_lstm_inv, "LSTM")
```
